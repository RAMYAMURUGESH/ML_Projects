#### Columns which has less than 8%-10 % missing value we will fill it with a median bcoz of outliers :)

```python
## to find percent missing:

def percent_df(df):
    a = df.columns[df.isna().any()].to_list()
    less_than_1percent, more_than_1percent = [],[]
    for i in a:
        percent  = 100*df[i].isna().sum()/len(df)
        if percent > 1 and percent<10:
            more_than_1percent.append(i)
        else:
            less_than_1percent.append(i)
            
    return more_than_1percent,less_than_1percent
```

```python
# to fill the statistical values for missing data which is more than 1%

def fill_median_mode(more_than_1percent,df):
    
    for i in more_than_1percent:
        
        if df[i].dtypes == 'object':
            m= df[i].mode()
            print(f'mode : {m}')
            df[i] = df[i].fillna(value=m,inplace=True)
        else :
           
            mm = df[i].median()
            print(f'median : {mm}')
            df[i] = df[i].fillna(value=mm,inplace=True)
    
    return df
```

```python
# to fill the dummy values for missing data which is less than 1%

def fill_dummy(less_than_1percent,df):
    for i in less_than_1percent:
        if df[i].dtypes == 'object':
            df[i] = df[i].fillna(value = 'None')
        else :
            df[i] = df[i].fillna(value = 0)
    return df
```

```python
more_than_1percent,less_than_1percent = percent_df(df)
print(f'more_than_1percent : {more_than_1percent} ; less_than_1percent :{less_than_1percent}')
```

```python
df = fill_median_mode(more_than_1percent,df)
df = fill_dummy(less_than_1percent,df)
```

```python
x = pd.get_dummies(df.drop('SalePrice',axis =1))
y = df['SalePrice']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import ElasticNet
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_absolute_error,mean_squared_error
import warnings
warnings.simplefilter('ignore')


x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.10,random_state=101)
scaler = StandardScaler()
scale_x_train = scaler.fit_transform(x_train)
scale_x_test = scaler.transform(x_test)
elastic_model = ElasticNet()

param_grid = {'alpha':[0.1,1,5,10,50,100],'l1_ratio':[.1, .5, .7, .9, .95, .99, 1]}
grid = GridSearchCV(estimator=elastic_model,param_grid=param_grid,scoring='neg_mean_squared_error',cv=5)
grid.fit(scale_x_train,y_train)
y_pred = grid.predict(scale_x_test)
residual = y_test - y_pred

print(f'Best Params : {grid.best_params_}')
print(f'Mean Absolute Error : {mean_absolute_error(y_pred,y_test)}')
#print(f'Mean Squared Error : {mean_squared_error(y_pred,y_test)}')
print(f'Root Mean Squared Error : {np.sqrt(mean_squared_error(y_pred,y_test))}')
```
